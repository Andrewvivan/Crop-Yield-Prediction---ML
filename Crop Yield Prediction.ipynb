```python
import pandas as pd
import numpy as np
import os
import glob
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
# Reading data
df_train = pd.read_csv("CropTrain2.csv")
df_test = pd.read_csv("CropTrain1.csv")

# Dropping unnecessary columns
df_train = df_train.drop(['Unnamed: 0'], axis=1)
df_test = df_test.drop(['Unnamed: 0'], axis=1)

# Separating target variable
y_train = df_train['Production']
y_test = df_test['Production']
df_train = df_train.drop(['Production'], axis=1)
df_test = df_test.drop(['Production'], axis=1)
```

```python
# Encoding categorical variables
cr = LabelEncoder()
se = LabelEncoder()
df_train['Season'] = se.fit_transform(df_train['Season'])
df_train['Crop'] = cr.fit_transform(df_train['Crop'])
df_test['Season'] = se.transform(df_test['Season'])
df_test['Crop'] = cr.transform(df_test['Crop'])

# Extracting values
x_train = df_train.values
x_test = df_test.values

# scaling data:
scaler = StandardScaler()
x_trainscaled = scaler.fit_transform(x_train)
x_testscaled = scaler.transform(x_test)
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error

# Linear Regression on non-scaled data
clf_linear = LinearRegression()
clf_linear.fit(x_train, y_train)
ypredlinear = clf_linear.predict(x_test)
maelinear = mean_absolute_error(y_test, ypredlinear)
r2scorelinear = r2_score(y_test, ypredlinear)
mselinear = mean_squared_error(y_test, ypredlinear)

# Linear Regression on scaled data
clf_linearscaled = LinearRegression()
clf_linearscaled.fit(x_trainscaled, y_train)
ypredlinearscaled = clf_linearscaled.predict(x_testscaled)
maelinearscaled = mean_absolute_error(y_test, ypredlinearscaled)
r2scorelinearscaled = r2_score(y_test, ypredlinearscaled)
mselinearscaled = mean_squared_error(y_test, ypredlinearscaled)

# Collect metrics for plotting
linear_mae_list = [maelinear, maelinearscaled]
linear_r2_list = [r2scorelinear, r2scorelinearscaled]
linear_mse_list = [mselinear, mselinearscaled]
namerandom = ['Non Scaled', 'Scaled']

# Plot MAE
plt.figure(figsize=(18, 5))
plt.subplot(1, 3, 1)
plt.bar(namerandom, linear_mae_list, color='skyblue')
plt.xlabel("Data")
plt.ylabel("Mean Absolute Error")
plt.title("Linear Regression MAE Comparison")

# Plot R² Score
plt.subplot(1, 3, 2)
plt.bar(namerandom, linear_r2_list, color='lightgreen')
plt.xlabel("Data")
plt.ylabel("R² Score")
plt.title("Linear Regression R² Score Comparison")

# Plot MSE
plt.subplot(1, 3, 3)
plt.bar(namerandom, linear_mse_list, color='salmon')
plt.xlabel("Data")
plt.ylabel("Mean Squared Error")
plt.title("Linear Regression MSE Comparison")

plt.tight_layout()
plt.show()
```
